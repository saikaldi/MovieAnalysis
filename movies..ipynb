```python
# importing all necessary libriaries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px
import plotly.io as pio
import seaborn as sns
```

```python
# Reading data from movies.csv
df = pd.read_csv('movies.csv')
df.head(3)
```

```python
df.shape
```

```python
df.columns
```

```python
# Checking null values
df.isnull().sum()
```

```python
# Drop Unnecessary Columns from the Dataset
df.drop(columns = ['id', 'imdb_id', 'homepage', 'cast', 'tagline', 'overview', 'budget_adj'], inplace=True)
```

```python
# Drop Unnecessary Columns from the Dataset
df.drop(columns = ['id', 'imdb_id', 'homepage', 'cast', 'tagline', 'overview', 'budget_adj'], inplace=True)
```

```python
# Check for missing values in each column of the DataFrame
df.isnull().sum()
```

```python
# Removing unnecessary rows 
df.dropna(how = 'any', subset = ['genres', 'director'], inplace=True) 
```

```python
# Replace NaN values with 0 in 'production_companies' and 'keywords' columns
df['production_companies'] = df['production_companies'].fillna(0)
df['keywords'] = df['keywords'].fillna(0)
```

```python
# Round the 'popularity' column values to 2 decimal places
df['popularity']=df['popularity'].round(2)
df
```

```python
# Inserting two columns profit and roi
df.insert(3, 'profit', df.revenue-df.budget)
df.insert(4, 'roi', df.profit / df.budget)
df
```

```python
# Round the 'roi' column values to 2 decimal places
df['roi']=df['roi'].round(2)
df
```

```python
# Select specific columns for further analysis
df1 = df[['popularity', 'budget', 'revenue', 'profit', 'roi', 'vote_count', 'vote_average', 'release_year']]
```

```python
# Check for missing values in each column of the DataFrame
df.isnull().sum()
# Display the count of unique values in the 'roi' column
df.roi.value_counts()
```

```python
# Identify non-finite values (NaN, inf) in the 'roi' column
non_finite_values = ~np.isfinite(df['roi'])
```

```python
# Count the total number of non-finite values (NaN or inf) in the 'roi' column
non_finite_values.sum()
```

```python
# Replace positive and negative infinity values in the 'roi' column with NaN
df['roi'] = df['roi'].replace([np.inf, -np.inf], np.nan)
```

```python
# Create histograms for each numerical column in df1 with 20 bins and display the plot with a size of 14x12 inches
df1.hist(bins =20, figsize = (14, 12))
plt.show()
```

```python
# Get the count of unique values in the 'popularity' column
df.popularity.value_counts
```

```python
# Group the data by 'release_year' and calculate the mean ROI for each year
df2=df.groupby('release_year')['roi'].mean()
# Plot the average ROI over different release years as a line plots
df2.plot(kind='line')
```

```python
# Group the data by 'release_year' and calculate the total popularity for each year
df3 = df.groupby('release_year')['popularity'].sum()
# Plot the total popularity over different release years as a line plot with red color
df3.plot(kind='line', color='red')
# Set labels for the x and y axes
plt.xlabel('Year', fontsize=12)
plt.ylabel('Popularity')
```

```python
# Group the data by 'release_year' and calculate the average vote rating for each year
df4 = df.groupby('release_year')['vote_average'].mean()
# Plot the average vote rating over different release years as a line plot
df4.plot(kind='line')
# Set labels for the x and y axes
plt.xlabel('Year', fontsize=12)
plt.ylabel('Rating')
```

```python
# Create a scatter plot showing the relationship between 'popularity' and 'vote_average'
df5 = df.plot.scatter(x = 'popularity', y = 'vote_average', c = 'green', figsize=(6,4))
# Set the x-axis label with a dark red color
df5.set_xlabel('Popularity', color = 'DarkRed')
# Set the y-axis label with a dark red color
df5.set_ylabel('Rating', color = 'DarkRed')
# Set the title of the plot with a font size of 17
df5.set_title('Popularity vs Rating', fontsize=17)
```

```python
# Visualizing
df8.plot(kind = 'bar', x = 'extracted_month', y = 'popularity', color ='DarkBlue')
```
